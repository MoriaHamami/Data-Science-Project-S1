# Web-Scraping from IMBD

First of all, let's scrape our data from IMBD:

Our code scrapes each page individually, therefore we wrote it four times to get more data.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

name=[]
rating=[]
genre=[]
year=[]
votes=[]
runtime=[]

url = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
data = requests.get(url).content
soup = BeautifulSoup(data,"html.parser")

for x in soup.findAll('div', attrs={'class':'lister-item mode-advanced'}):
    header = x.find('h3', attrs={'class':'lister-item-header'})
    temp_name=header.find('a', href=True)
    temp_rating=x.find('div', attrs={'class':'inline-block ratings-imdb-rating'})
    temp_genre=x.find('span', attrs={'class':'genre'})
    temp_year=x.find('span', attrs={'class':'lister-item-year text-muted unbold'})
    temp_votes=x.find('span', attrs={'name':'nv'})
    temp_runtime=x.find('span', attrs={'class':'runtime'})
    name.append(temp_name.text)
    rating.append(temp_rating.text.strip("\n"))
    votes.append(temp_votes.text)
    runtime.append(temp_runtime.text.strip("\n"))
    genre.append(temp_genre.text.strip("\n"))
    year.append(temp_year.text)

data_collection=pd.DataFrame({'Title':name, 'Rating':rating, 'Votes' :votes, 'Genre':genre,'Years':year, 'Runtime':runtime})
data_collection.to_csv('DS-PR1-18IT102.csv', index=False, encoding="utf-8-sig")
```
